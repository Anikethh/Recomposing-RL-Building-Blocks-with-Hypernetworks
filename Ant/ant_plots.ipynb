```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

def plot_configuration(ax, rewards, std_deviations, label):
    x = np.arange(len(rewards))
    ax.plot(x, rewards, label=label)
    ax.fill_between(x, rewards - std_deviations, rewards + std_deviations, alpha=0.2)

# Assuming you have multiple configurations, load data for each configuration
configurations = [
    {"label": "MLP", "rewards_file": "ant_mlp_100k\\rewards_AntBulletEnv-v0_seed5.npy", "std_devs_file": "ant_mlp_100k\std_devs_AntBulletEnv-v0_seed5.npy"},
    {"label": "Hyper", "rewards_file": "ant_hyper_100k\\rewardsHyper__Walker2DBulletEnv-v0_seed5.npy", "std_devs_file": "ant_hyper_100k\std_devsHyper__Walker2DBulletEnv-v0_seed5.npy"},
    {"label": "Hyperfan Prioritised", "rewards_file": "ant_hyperfan_100k\\rewards_Prioritised_Hyperfan_AntBulletEnv-v0_seed5.npy", "std_devs_file": "ant_hyperfan_100k\std_devs_Prioritised_Hyperfan_AntBulletEnv-v0_seed5.npy"},
    {"label": "Dual Actor", "rewards_file": "ant_hyperfan_dualactor_100k\\rewards_Hyperfan_DualActor_AntBulletEnv-v0_seed5.npy", "std_devs_file": "ant_hyperfan_dualactor_100k\std_devs_Hyperfan_DualActor_AntBulletEnv-v0_seed5.npy"}
]

# Create subplots for each configuration
fig, ax = plt.subplots(figsize=(10, 5))
sns.set()

for config in configurations:
    rewards = np.load(config["rewards_file"])
    std_deviations = np.load(config["std_devs_file"])
    plot_configuration(ax, rewards, std_deviations, config["label"])

# Add legends, labels, and show the plot
ax.legend()
ax.set_xlabel('Evaluation Steps')
ax.set_ylabel('Mean Reward')

plt.tight_layout()
plt.show()
```
