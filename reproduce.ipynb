```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

def plot_configuration(ax, rewards, std_deviations, label):
    x = np.arange(len(rewards))
    ax.plot(x, rewards, label=label)
    ax.fill_between(x, rewards - std_deviations, rewards + std_deviations, alpha=0.2)

# Load Walker rewards and standard deviations
rewards_walker = np.concatenate([rewardsw0, rewardsw1, rewardsw2, rewardsw3])
std_devs_walker = np.concatenate([std_devsw0, std_devsw1, std_devsw2, std_devsw3])

# Load Ant rewards and standard deviations
rewards_ant = np.concatenate([rewardsa0, rewardsa1, rewardsa2, rewardsa3])
std_devs_ant = np.concatenate([std_devsa0, std_devsa1, std_devsa2, std_devsa3])

# Load Hopper rewards and standard deviations
rewards_hopper = np.load("D:\IIIT\Sem7\RL\Project\Reproduced\Hopper\\rewards_Hyperfan_DualActor__HopperBulletEnv-v0_seed5.npy")
std_devs_hopper = np.load("D:\IIIT\Sem7\RL\Project\Reproduced\Hopper\std_devs_Hyperfan_DualActor__HopperBulletEnv-v0_seed5.npy")

# Create subplots for each environment (Walker, Ant, and Hopper)
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(18, 5))
sns.set()

# Plot Walker
plot_configuration(ax1, rewards_walker, std_devs_walker, 'Walker')
ax1.legend()
ax1.set_xlabel('Evaluation Steps')
ax1.set_ylabel('Mean Reward')
ax1.set_title('Walker Environment')

# Plot Ant
plot_configuration(ax2, rewards_ant, std_devs_ant, 'Ant')
ax2.legend()
ax2.set_xlabel('Evaluation Steps')
ax2.set_ylabel('Mean Reward')
ax2.set_title('Ant Environment')

# Plot Hopper
plot_configuration(ax3, rewards_hopper, std_devs_hopper, 'Hopper')
ax3.legend()
ax3.set_xlabel('Evaluation Steps')
ax3.set_ylabel('Mean Reward')
ax3.set_title('Hopper Environment')

plt.tight_layout()
plt.show()
```
